```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
x = [0, 0.8975979, 1.7951958, 2.692794, 3.5903916, 4.48798951, 5.3855874, 6.2831853]
y = [1.00000, 0.52103, -0.15540, -0.52580, -0.43939, -0.09069, 0.21235, 0.28461]

cs = CubicSpline(x, y, bc_type='natural')

x_interp = np.linspace(min(x), max(x), 1000)
y_interp = cs(x_interp)

plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_interp, y_interp, label='Cubic Spline')
plt.legend()
plt.show()
```

```python
x_a = 3
y_a = cs(x_a)
y_true_a = np.exp(-0.2 * x_a) * np.cos(x_a)
error_a = abs(y_a - y_true_a) / y_true_a
print("Valor da função spline aproximada em x = 3: ", y_a)
print("Erro relativo em relação ao f(x) original em x = 3: ", error_a)

x_b = 3.5903916
y_b = cs(x_b, 1)
y_prime_b = cs(x_b, 1)
y_prime_prime_b = cs(x_b, 2)
y_true_b = np.exp(-0.2 * x_b) * np.cos(x_b)
y_true_prime_b = -0.2 * np.exp(-0.2 * x_b) * np.cos(x_b) - np.exp(-0.2 * x_b) * np.sin(x_b)
y_true_prime_prime_b = -0.4 * np.exp(-0.2 * x_b) * np.cos(x_b) + 0.2 * np.exp(-0.2 * x_b) * np.sin(x_b)
error_b = abs(y_b - y_true_b) / y_true_b
error_prime_b = abs(y_prime_b - y_true_prime_b) / y_true_prime_b
error_prime_prime_b = abs(y_prime_prime_b - y_true_prime_prime_b) / y_true_prime_prime_b
print("Valor da função spline aproximada em x = 3.5903916: ", y_b)
print("Valor da primeira derivada da função spline aproximada em x = 3.5903916: ", y_prime_b)
print("Valor da segunda derivada da função spline aproximada em x = 3.5903916: ", y_prime_prime_b)
print("Erro relativo da função spline aproximada em relação ao f(x) original em x = 3.5903916: ", error_b)
print("Erro relativo da primeira derivada da função spline aproximada em relação ao f(x) original em x = 3.5903916: ", error_prime_b)
print("Erro relativo da segunda derivada da função spline aproximada em relação ao f(x) original em x = 3.5903916: ", error_prime_prime_b)
```
